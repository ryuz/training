```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# ホワイトノイズを作成
x =  np.random.normal(0, 1.0, 1000)
X = np.fft.fft(x)

# 表示
fig = plt.figure(figsize=(16, 4))
ax_x = fig.add_subplot(121)
ax_x.plot(x)
ax_X = fig.add_subplot(122)
ax_X.plot(np.abs(X))
plt.show()
```

```python
# 移動平均フィルタを掛けてみる
x1 = x.copy()
for i in range(1, len(x)):
    x1[i] = (x[i-1] + x[i]) / 2

# FFT
X1 = np.fft.fft(x1)

# 表示
fig = plt.figure(figsize=(16, 4))
ax_x = fig.add_subplot(121)
ax_x.plot(x1)
ax_X = fig.add_subplot(122)
ax_X.plot(np.abs(X1))
plt.show()
```

```python
# 差分フィルタを掛けてみる
x1 = x.copy()
for i in range(1, len(x)):
    x1[i] = (x[i-1] - x[i]) / 2

# FFT
X1 = np.fft.fft(x1)

# 表示
fig = plt.figure(figsize=(16, 4))
ax_x = fig.add_subplot(121)
ax_x.plot(x1)
ax_X = fig.add_subplot(122)
ax_X.plot(np.abs(X1))
plt.show()
```
