```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

KLダイバージェンス(確率分布間の距離)を試してみる

$$
\begin{align}
KL(p||q) &= - \int{p(x) \ln{q(x)} dx} - \left(- \int{p(x) \ln{p(x)} dx} \right) \\
         &= - \int{p(x) \ln{ \left( \frac{q(x)}{p(x)} \right) }  } dx
\end{align}
$$

```python
def KL(p, q):
    return -np.sum(p*np.log(q/p))
```

実験の対象に正規分布を使ってみる

```python
from scipy.stats import norm
x = np.linspace(-5.0, 5.0, 1000)
y0 = norm.pdf(x, loc=0, scale=1)
```

```python
plt.plot(x, y0)
```

## 正規分布の平均値が変わると結果がどう変わるか？

```python
u = np.linspace(-5.0, +5.0, 100)
l = []
for uu in u:
    l.append(KL(y0, norm.pdf(x, loc=uu, scale=1)))
plt.plot(u, l)
```

## 正規分布の標準偏差が変わると結果がどう変わるか？

```python
s = np.linspace(0.5, 4.0, 100)
l = []
for ss in s:
    l.append(KL(y0, norm.pdf(x, loc=0, scale=ss)))
plt.plot(s, l)
```
