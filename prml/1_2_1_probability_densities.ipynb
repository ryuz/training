# 1.2.1 確率密度

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
def gaussian(x, u=0., s=1.):
    return 1/(np.sqrt(2*np.pi)*s) * np.exp(-(x - u)**2 / (2*s**2))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

適当な関数fを定義して(とりあえずガウス関数)、非線形な関数gで変数変換してみる(とりあえずシグモイド)

```python
def Px(x):
    """xの確率分布"""
    return gaussian(x, 1, 1)

def g(y):
    """変数変換関数g"""
    return sigmoid(y)

def gg(y):
    """gの微分"""
    return (1 - sigmoid(y))*sigmoid(y)
```

```python
x = np.linspace(-5, +5, 1000)
plt.title('Px(x)')
plt.plot(x, Px(x))
plt.show()

y = np.linspace(-5, +5, 1000)
plt.figure(figsize=(8,3))
plt.subplot(121)
plt.title('g(y)')
plt.plot(y, g(y))
plt.subplot(122)
plt.title("g'(y)")
plt.plot(y, gg(y))
plt.show()
```

$ x = g(y) $ とすると y の確率分布 $ P_y $ は

$ P_y(y) = p_x(x) | \frac{dx}{dy} | = p_x(g(y)) | g'(y) |$

```python
y = np.linspace(-5, +5, 1000)
plt.title('Py(y)')
plt.plot(y, Px(g(y))*np.abs(gg(y)))
```
