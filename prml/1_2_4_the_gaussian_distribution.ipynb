# 1.2.4 ガウス分布

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn
plt.style.use('ggplot')
%matplotlib inline
```

## サンプル数と推定

図1.5に出てくる話をサンプル数を変えてやってみる

サンプル数と推定誤差を何となく見た目で体感してみる

```python
# サンプル数10でやってみる
n = 5
x = np.linspace(-5, +5, 100)
plt.figure(figsize=(12, 4))
for i in range(10):
    plt.subplot(2, 5, i+1)
    z = np.zeros((n))
    y = np.random.normal(0, 1, size=n)
    u = np.mean(y)
    s = np.std(y)
    plt.title('n={}'.format(n))
    plt.scatter(y, z, color='green')
    plt.plot(x, norm.pdf(x, 0, 1))
    plt.plot(x, norm.pdf(x, u, s))
plt.tight_layout()
```

## 尤度マップ書いてみる

この式をで尤度をヒートマップを書いてみる(横軸 $ \mu $, 縦軸 $ \sigma $ )

$$
p(\boldsymbol{t}|\boldsymbol{x},\boldsymbol{w},\beta) = \prod_{n=1}^{N} N(t_n|y(x_n,\boldsymbol{w},\beta^{-1*})
$$

```python
n = 5
y = np.random.normal(0, 1, size=n)
print('mean = {}'.format(np.mean(y)))
print('std  = {}'.format(np.std(y)))

u = np.linspace(-2, +2, 100)
s = np.linspace(0.01, +5, 100)
likely_map = np.ndarray((len(s), len(u)))
for i in range(len(u)):
    for j in range(len(s)):
        likely_map[i][j] = np.prod(norm.pdf(y, loc=u[i], scale=s[j]))
seaborn.heatmap(likely_map)
```

結局のところ最尤推定すると $ \mu $ と $ \sigma $ なのだけど広がりを見るのは面白い
